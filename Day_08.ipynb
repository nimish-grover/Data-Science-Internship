```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X=2*np.random.rand(100,1)
Y=4+3*X + np.random.randn(100,1)
```

```python
#add a bias term to X
#X_b is the input matrix with an additional bias term. we initialize the model 
X_b=np.c_[np.ones((100,1)),X]
```

```python
#initialize the model parameter
theta=np.random.randn(2,1)
theta
```

```python
# hyperparameters
learning_rate= 0.1
n_iterations = 1000
```

```python
#gradient descent
for iteration in range(n_iterations):
    gradients=2*X_b.T.dot(Y-X_b.dot(theta))
    theta-=learning_rate *gradients
theta
```

```python
#optimal parameter
intercept,slope=theta
print("Intercept:",intercept[0])
print("slope:",slope[0])
```

Unsupervised learning



```python
#generate synthetic data 
np.random.seed(42)
X = np.random.rand(100,1)*10
Y = 2 * X + 3 + np.random.randn(100,1)
```

```python
#linear regression model class
class LinearRegression:
    def __init__(self):
        self.weights = None
        self.bias = None

    def fit(self,X,Y,learning_rate=0.01,epochs=100):
        num_samples,num_features=X.shape
        self.weights=np.zeros((num_features,1))
        self.bias=0

        for _ in range(epochs):
            y_pred = np.dot(X, self.weights) + self.bias
            dw=(1/num_samples) * np.dot(X.T,(y_pred-Y))
            db=(1/num_samples) * np.sum(y_pred-Y)

            self.weights-=learning_rate * dw
            self.bias -=learning_rate * db

    def predict(self,X):
        return np.dot(X,self.weights)+self.bias
```

```python
#instantiate and train model 
model = LinearRegression()
model.fit(X,Y, learning_rate = 0.01, epochs = 20)
x_test = np.array([[1],[5],[10]])
y_pred = model.predict(x_test)
```

```python
#plotting the graph
plt.scatter(X,Y,label ='Training Data')
plt.plot(x_test,y_pred, color="red",label = "Linear Regression")
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Linear Regression using Numpy")
plt.legend()
plt.show()
```
