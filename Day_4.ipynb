```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
df=pd.read_csv('Pokemon.csv')
df
```

```python
#coloumn operations
print(df.head(n=10))
df.columns = df.columns.str.upper().str.replace('_','')
#change into upper case
df.head()
```

```python
df[df['LEGENDARY']==True].head(5)
#showing legendary pokemon
df
```

```python
df=df.set_index('NAME')
#change and set the index to the name attribute
df
```

```python
df.index = df.index.str.replace(".*(?=Mega)",'')
df.head(10)
df
```

```python
df=df.drop(["#"],axis=1)
#drop the coloumns with axis 1 axis 0 is for rows
print("the coloumns of the dataset are",df.columns)
#show the dataframes
print("the shape of dataframe is:",df.shape)
#shape of the dataframe
```

```python
print(type(df.index))
print(df[((df['TYPE 1']=='Water')|(df['TYPE 1']=="Dragon"))&((df['TYPE 2']=='Dragon')
                                                             |(df['TYPE 2']=='Fire'))].head(3))
print("MAX HP:",df['HP'].argmax())
#returns pokemon with highest hp
```

```python
print("MAX DEFENSE:",(df['DEFENSE']).idxmax())
#similar to argmax
```

```python
df.sort_values('TOTAL',ascending=False).head(3)
#this arranges pokemon in descending order of the totals
#sort_values is used for sorting and ascending-Flase is making it in descending order
```

```python
df2=df.drop(['GENERATION','TOTAL'],axis=1)
#here we will be plotting with seaborn
sns.boxplot(data=df2)
#setting the box size 
plt.ylim(0,300)
#change the scale of y axis
plt.show()
```
